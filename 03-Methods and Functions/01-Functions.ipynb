Check if any number in a list is even

```python
def even_check_list(num_list):
   
    for num in num_list:
        if num % 2== 0:
            return True
        
    
    return  False;
```

```python
even_check_list([1,3,5])
```

```python
even_check_list([2,4,1])
```

```python
work_hours = [('Abby',100),('Billy',4000),('Cassie',800)]
```

```python
def employee_check(work_hours):
    max = 0
    emp = ''

    for name, hours in work_hours:
        if hours > max :
            max = hours
            emp = name
    # return
    return (emp,max)
    
```

```python
print (employee_check(work_hours))
```

<b>Interactions between functions </b>

```python
example = [1,2,3,4,5]
```

```python
from random import shuffle
shuffle(example)
```

```python
example
```

```python
def shuffle_list(mylist):
    shuffle(mylist)
    return mylist
```

```python
result = shuffle_list(example)
```

```python
result
```

```python
def player_guess():
    guess = ''
    while guess not in ['0','1','2']:
        guess  = input ('pick no')

    return int (guess)
    
```

```python
def check_guess(mylist, guess):
    if mylist[guess] == 'O':
        print('Correct')
    else: 
        print('Wrong')
        print(mylist)
        
    
```

```python
mylist = ['O', '', '']
mixedup_list = shuffle_list(mylist)
guess = player_guess()
check_guess(mixedup_list,guess)
```
