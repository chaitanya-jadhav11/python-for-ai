```python
def myFunc(a,b): # fixed no of parameter
    return (a+b) * 0.05
```

```python
myFunc(40,60)
```

```python
def myFunc(*args): # dyanamic parameter. it create Tulip data type
    print(args) # print
    for item in args: # print
        print(item)
    return sum(args) * 0.05
```

```python
myFunc(40,60)
```

```python
myFunc(40,60,80)
```

```python
def myFunc2(*args,**kargs ): # dyanamic parameter. it create dict data type
    print(args) # print
    print(kargs) # print
```

```python
myFunc2(10,20,30,fruits='Apple', veggi='bhendi')
```

```python
def myfunc(*args):
    mtlist = []
    for i in args:
        if i % 2 ==0 :
            mtlist.append(i)
    return mtlist
```

```python
print (myfunc(1,2,3,4,5,6))
```
