String Formatting

```python
str='String'
```

```python
print('This is a string {}'.format('INSERT'))
```

```python
print('This is a string {} {}'.format('INSERT', 'HELLO'))
```

```python
print('This is a string {1} {0}'.format('INSERT', 'HELLO'))
```

```python
print('This is a string {f} {d}'.format(f=str, d='HELLO'))
```

Float formatting

```python
result = 100/777
```

```python
result
```

```python
print("result {r} ".format(r=result))
```

```python
print("result {r:10.3f} ".format(r=result))
```

```python
print("result {r} ".format(r=result))
```
